1. Why are functions advantageous to have in your programs?

Ans -  Functions are reusable piece of code used in programs. It helps to re-use the same logic multiple times in the same program and avoid duplication.


2. When does the code in a function run: when it's specified or when it's called?
    
Ans - Its run when it is called. If the function used is not called, then all the contents defined under the function is not executed.    
      
3. What statement creates a function?

Ans - Functions are created using the def keyword.
      Eg : def add(a,b):
               return a + b
            

4. What is the difference between a function and a function call?

Ans - Function is code which does all the computation defined under it and returns a result whereas function call invokes the execution
      of the function if its passed with the same arguments as defined in the function.
    

5. How many global scopes are there in a Python program? How many local scopes?

Ans - There is one global scope applicable to the whole program. Local scopes is the scope defined under each function in the program
       and we can have multiple local scopes in the whole program.
    
    
6. What happens to variables in a local scope when the function call returns?

Ans - Variables in local scope is removed when the function exceution ends. 


7. What is the concept of a return value? Is it possible to have a return value in an expression?

Ans - Return value is returned when the function execution ends. If its specified in the function with the return keyword, then it
      returns that value else None. Yes , its possible to hava a return value in an expression.


8. If a function does not have a return statement, what is the return value of a call to that function?

13. What can you do to save a programme from crashing if it encounters an error?

Ans - We can used try-except block. The error message can be defined under the except block. 

14. What is the purpose of the try clause? What is the purpose of the except clause?

Ans -  All the statements are executed in a normal way under try clause. But if there is an unexpected error  while executing
       the statements under try block, the control moves to the Except block and its skips the code in between the try and
       except block. Except block is used to catch unexpected errors and it helps to avoid the crashing of the whole program.
